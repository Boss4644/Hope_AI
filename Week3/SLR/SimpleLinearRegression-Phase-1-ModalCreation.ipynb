```python
#import pandas library which is used to read data from any type documents
import pandas as pd
```

```python
#read data csv file and stored in dataset variable
dataset = pd.read_csv("Salary_Data.csv") 
```

```python
#get YearsExperience(input) column data from dataset variable and assigned into independent variable
independent = dataset[["YearsExperience"]]
```

```python
#get Salary(output) column data from dataset variable and assigned into dependent variable
dependent = dataset[["Salary"]]
```

```python
#split train and test set using sklearn.model_selection library with train_test_split function
from sklearn.model_selection import train_test_split
#independent - input
#dependent - output
#test_size - test data set size - 0.3(30% data split as test data set)
#random_state - not yet explained
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
#X_train - input, Y_train - output (for training)
#X_test - input, Y_test - output (for testing)
```

```python
#import sklearn.linear_model library with LinearRegression class for find linear regression formula or create model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
#fit method is used to create model with training input and output
regressor.fit(X_train, Y_train)
```

```python
#get weight and bias
regressor.coef_ #weight
regressor.intercept_ #bias
```

```python
#get prediction value with test input
y_pred = regressor.predict(X_test) 
```

```python
#metric evaluation using sklearn.metrick library with r2_score function
from sklearn.metrics import r2_score
#get r_square value using test output and predicted output
r_score = r2_score(Y_test, y_pred)
#r_score value must be between 0 and 1, if its nearly 1 then its good model, if its nearly 0 then its bad model
r_score
```

```python
#save model using pickle library
import pickle
#define model name
file_name = "finalized_model_linear.sav"
#save model with defined model name using dump function
pickle.dump(regressor, open(file_name, "wb"))
```

```python
#use model in project using pickle library
loaded_model = pickle.load(open(file_name, "rb"))
result = loaded_model.predict([[13]])
```
